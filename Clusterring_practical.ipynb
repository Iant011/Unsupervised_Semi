```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
df
```

```python
x = df.iloc[:,3:]
```

```python
x
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = [] 
for i in range(1,15):
    kmean = KMeans(n_clusters=i,init='k-means++',random_state=30)
    kmean.fit(x)
    wcss.append(kmean.inertia_) # summation of squares
```

```python
wcss
```

```python
plt.plot(range(1,15),wcss)
```

```python
kmean1 = KMeans(n_clusters=5,init='k-means++',random_state=30)
kmean1.fit_predict(x)
```

```python
x['Clusters'] = kmean1.fit_predict(x)
```
